# Building a nicely-formatted emotion library for portuguese
I will use the LIWC Portuguese Dictionary, avaliable [here](http://143.107.183.175:21380/portlex/index.php/pt/projetos/liwc).
It has a very big corpus and measures 64 emotions/categories related to the words.
In order to use it, I will try formatting it in the same way the english [EmotionLexicon](http://saifmohammad.com/WebPages/NRC-Emotion-Lexicon.htm) is formatted.

```python
import pandas as pd
```

```python
liwc = pd.read_csv('LIWC_port.txt',encoding='Latin5')
```

```python
liwc.head()
```
